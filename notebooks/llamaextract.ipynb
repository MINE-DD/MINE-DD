# llama extract Resumes example


Example on how to use llama_extract with Pydantic. Original example version and data obtained from: [run-llama /llama-extract](https://github.com/run-llama/llama_extract/blob/main/examples/demo_analysis.ipynb)

```python
from llama_extract import LlamaExtract
import os
import nest_asyncio

nest_asyncio.apply()
```

```python
os.environ["LLAMA_CLOUD_API_KEY"] = input("please provide your llama cloud key")
```

```python
DATA_DIR = #
```

```python
fnames = os.listdir(DATA_DIR)
fnames = [fname for fname in fnames if fname.endswith(".pdf")]
fpaths = [os.path.join(DATA_DIR, fname) for fname in fnames]
fpaths
```

```python
extractor = LlamaExtract(base_url=None, check_interval=1, max_timeout=1000, num_workers=4, show_progress=True, verbose=False)
```

```python
fpaths_for_infer_schema = fpaths[:1]
fpaths_for_infer_schema
```

```python
extraction_schema = await extractor.ainfer_schema(
    "TestSchemaResumes", seed_files=fpaths_for_infer_schema, project_id=None
)
```

```python
extraction_schema.data_schema
```

```python
extractions = await extractor.aextract(
    files=
    ["../../Data/testdata/noisebridge_receipt.pdf", "../../Data/testdata/parallels_invoice.pdf"], project_id=None, response_model=None, schema_id=extraction_schema.id
)
```

```python
extractions = await extractor.aextract(schema_id=extraction_schema.id, files=["../../Data/testdata/noisebridge_receipt.pdf", "../../Data/testdata/parallels_invoice.pdf"], project_id=None, response_model=None)
```

```python
extraction = extractor.extract(schema_id=extraction_schema.id, file_input=["../../Data/testdata/noisebridge_receipt.pdf"], project_id=None, response_model=None)
```
