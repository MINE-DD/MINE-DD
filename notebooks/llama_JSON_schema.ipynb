# Using LlamaExtract with JSON Schema

Example on how to use llama_extract with Pydantic. Original version of the example obtained from: [run-llama /llama-extract](https://github.com/run-llama/llama_extract/blob/main/examples/demo_json_schema.ipynb)

[LLAMA 3.1 community license agreement](https://llama.meta.com/llama3_1/license/)

```python
import os
from llama_extract import LlamaExtract
```

```python
os.environ["LLAMA_CLOUD_API_KEY"] = input("please provide your llama cloud key")
```

```python
extractor = LlamaExtract(base_url=None, check_interval=1, max_timeout=2000, num_workers=4, show_progress=True, verbose=True)

data_schema = {
    "type": "object",
    "properties": {
        "number": {"type": "string"},
        "date": {"type": "string"},
        "amount": {"type": "number"},
    },
}

```

```python
extraction_schema = await extractor.acreate_schema("Json", data_schema)
```
